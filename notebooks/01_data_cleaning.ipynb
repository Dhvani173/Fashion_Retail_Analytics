```python
import pandas as pd

df = pd.read_csv("../data/raw/Fashion_Retail_Sales.csv")
df.head()
```

```python
df['Date Purchase'] = pd.to_datetime(df['Date Purchase'], format="%d-%m-%Y", errors='coerce')
df['Review Rating'] = df['Review Rating'].fillna(df['Review Rating'].median())
df.dropna(subset=['Purchase Amount (USD)'], inplace=True)
```

```python
df['Year'] = df['Date Purchase'].dt.year
df['Month'] = df['Date Purchase'].dt.month
df['Weekday'] = df['Date Purchase'].dt.day_name()
```

```python
df.columns = df.columns.str.strip().str.replace(' ', '_').str.lower()
df.head()
```

```python
df.to_csv("../data/processed/cleaned_fashion_sales.csv", index=False)
print("✅ Cleaned file saved!")
```
